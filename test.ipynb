```python
import torch
x = torch.randn(4)
torch.norm(x, p=1)
```

```python
import torch
x = torch.tensor([1.0], requires_grad=True)
y = x  # y 与 x 指向同一张量（浅拷贝）
y.add_(x)  # 原地修改 y（即 x）为 2.0（覆盖原始值 1.0）
loss = y.norm()  # loss = 2.0
loss.backward()
print(x.grad)  # 错误结果！
```

```python
import torch
x = torch.tensor([-1.0], requires_grad=True)
y = torch.relu(x)  # y = 0.0（因 x < 0），grad_fn 为 ReLUBackward0
y.add_(x)  # 原地修改 y 为 -1.0（覆盖原始值 0.0）
loss = y.norm()
loss.backward()
print(x.grad)  # 错误结果！
```
